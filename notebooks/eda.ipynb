```python
user_withdrawals = withdrawals_df.groupby("advertiser_zrive_id").agg(
    num_withdrawals=("withdrawal_id", "count"),
    withdrawal_types=("withdrawal_type", lambda x: ", ".join(x.unique()))
).sort_values("num_withdrawals", ascending=False)

user_withdrawals[user_withdrawals["num_withdrawals"] > 1].head(10)
```

```python
definitive_churn_condition = (
    (withdrawals_df["withdrawal_type"] == "TOTAL") &
    (withdrawals_df["withdrawal_status"] != "Denegada") &
    (withdrawals_df["withdrawal_reason"].isin([
        "Upselling-cambio de contrato",
        "Cambio a Bundle Online",
        "Cambio de Contrato/propuesta/producto"
    ]))
)
```

```python
definitive_withdrawals_count = (
    withdrawals_df[definitive_churn_condition]
    .groupby("advertiser_zrive_id")
    .size()
    .reset_index(name="definitive_withdrawals_count")
)

advertisers_multiple_definitive_withdrawals = definitive_withdrawals_count[
    definitive_withdrawals_count["definitive_withdrawals_count"] > 1
]

print(f"Number of advertisers with multiple definitive withdrawals: {len(advertisers_multiple_definitive_withdrawals)}")
advertisers_multiple_definitive_withdrawals.head()
```

```python
# Find advertisers with definitive withdrawals and check how many returned later

definitive_withdrawals = (
    withdrawals_df[definitive_churn_condition]
    .groupby("advertiser_zrive_id")
    .agg({"withdrawal_creation_date": "min"}) # Obtain min date
    .reset_index()
    .rename(columns={"withdrawal_creation_date": "definitive_withdrawal_date"})
)

merged_data = pd.merge(
    definitive_withdrawals,
    advertiser_df[["advertiser_zrive_id", "max_start_contrato_nuevo_date"]],
    on="advertiser_zrive_id",
)

merged_data["returned_advertisers"] = (
    merged_data["max_start_contrato_nuevo_date"] > merged_data["definitive_withdrawal_date"]
)

num_definitive_withdrawals = len(definitive_withdrawals)
num_returned_advertisers = merged_data["returned_advertisers"].sum()
porcentaje_retorno = (num_returned_advertisers / num_definitive_withdrawals) * 100

print(f"Advertisers with definitive withdrawal: {num_definitive_withdrawals}")
print(f"Advertisers who returned later: {num_returned_advertisers}")
print(f"Rate of return: {porcentaje_retorno:.2f}%")
```

```python
advertiser_df["min_start_contrato_date"] = pd.to_datetime(advertiser_df["min_start_contrato_date"], errors="coerce")
advertiser_df["max_start_contrato_nuevo_date"] = pd.to_datetime(advertiser_df["max_start_contrato_nuevo_date"], errors="coerce")
advertiser_df["contrato_churn_date"] = pd.to_datetime(advertiser_df["contrato_churn_date"], errors="coerce")
```

```python
# 1. Average and median time in new users
df_min = advertiser_df.dropna(subset=["min_start_contrato_date", "contrato_churn_date"]).loc[advertiser_df["max_start_contrato_nuevo_date"].isna()]
min_diffs = (df_min["contrato_churn_date"] - df_min["min_start_contrato_date"]).dt.days
min_mean = min_diffs.mean()
min_median = min_diffs.median()

print(f"Average time in new users: {int(min_mean)} days")
print(f"Median time in new users: {int(min_median)} days")
```

```python
# 2. Average and median time in old users
df_max = advertiser_df.dropna(subset=["max_start_contrato_nuevo_date", "contrato_churn_date"])
max_diffs = (df_max["contrato_churn_date"] - df_max["max_start_contrato_nuevo_date"]).dt.days
max_mean = max_diffs.mean()
max_median = max_diffs.median()

print(f"Average time in old users: {int(max_mean)} days")
print(f"Median time in old users: {int(max_median)} days")
```

```python
# 3. Average and median time combining new and old users
advertiser_df["initial_date"] = advertiser_df["max_start_contrato_nuevo_date"].fillna(advertiser_df["min_start_contrato_date"])
df_combined = advertiser_df.dropna(subset=["contrato_churn_date"])
combined_diffs = (df_combined["contrato_churn_date"] - df_combined["initial_date"]).dt.days
combined_mean = combined_diffs.mean()
combined_median = combined_diffs.median()

print(f"Average time: {int(combined_mean)} days")
print(f"Median time: {int(combined_median)} days")
```

```python
print(
    f"Metrics data is between: "
    f"{pd.to_datetime(metrics_df['period_int'], format='%Y%m').min().strftime('%Y-%m')} - "
    f"{pd.to_datetime(metrics_df['period_int'], format='%Y%m').max().strftime('%Y-%m')}"
)
```

```python
metrics_df["period_int"] = pd.to_datetime(metrics_df["period_int"], format="%Y%m")

monthly_trend = metrics_df.groupby("period_int")[
    ["monthly_contracted_ads", "monthly_published_ads", "monthly_total_invoice"]
].sum()

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 10), sharex=True)

axes[0].plot(monthly_trend.index, monthly_trend["monthly_contracted_ads"], label="Contracted Ads", marker="o")
axes[0].plot(monthly_trend.index, monthly_trend["monthly_published_ads"], label="Published Ads", marker="s")
axes[0].set_ylabel("Number of Ads")
axes[0].set_title("Monthly Contracted and Published Ads")
axes[0].legend()
axes[0].grid(True)

axes[1].plot(monthly_trend.index, monthly_trend["monthly_total_invoice"], label="Total Invoice", marker="^", color="red")
axes[1].set_ylabel("Total Invoice")
axes[1].set_title("Monthly Total Invoice")
axes[1].legend()
axes[1].grid(True)

plt.xticks(rotation=45)
plt.xlabel("Time (Months)")

plt.tight_layout()
plt.show()
```
